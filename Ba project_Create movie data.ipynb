```python
def create_columns(name,my_year):
    import re
    name1=name
    actor_se1=imdb.search_for_person(name1)
    #print(actor_se)
    #for i in range(len(actor_se1)):
        #if (actor_se1[i]["name"]==name1):
    actor_id1=actor_se1[0]["imdb_id"]
    url1 = "http://www.imdb.com/name/"+actor_id1+'/awards?ref_=nm_awd'
    response1 = requests.get(url1)
    if response1.status_code == 200:
        #print(response.content)
        page_data_soup1 = BeautifulSoup(response1.content,'lxml')
        #print(page_data_soup1.prettify())
    awardstype=page_data_soup1.find_all('h3')

    aw=[]
    for i in awardstype:
        aw.append(i.get_text())
    aw=aw[1:-3]
    awardstable=page_data_soup1.find_all('table')
    awardstable=awardstable[0:len(aw)]
    awy=[]
    w_n=[]
    a_type=[]
    awards_score=[]
    for i in awardstable:
        awardsyear=i.find_all('a')
        won_nom=i.find_all('b')
        award_type=i.find_all('span')
        for j in awardsyear:
            x=re.findall(r'\d+',j.get_text())
            if len(x)==1:
                awy.append(x[0])   
        for k in won_nom:
            if k.get_text()=="Nominated":
                w_n.append(1)
            elif k.get_text()=="Won":
                w_n.append(4)
        for l in award_type:
            x=re.findall(r'\d+',l.get_text())
            if len(x)==0:
                a_type.append(l.get_text())
    for m,n,p in zip(awy,w_n,a_type):
        awards_score.append((m,n,p))
    sum_1=0
    sum_2=0
    for a in awards_score:
        if a[2]=='Oscar':
            sum_1=sum_1+a[1]
            sum_2=sum_2+a[1]
        elif a[2]=='Golden Globe':
            sum_1=sum_1+a[1]
            sum_2=sum_2+a[1]
        elif a[2]=='BAFTA Film Award':
            sum_1=sum_1+a[1]
            sum_2=sum_2+a[1]
        else:
            sum_2=sum_2+a[1]
            
    #print(sum_1,sum_2)
    if sum_1>0:
        coeff=0.7*sum_2/sum_1
    else:
        coeff=0
    awards_score1=[]
    for a in awards_score:
        if a[2]=='Oscar':
            net_sc=a[1]*coeff
        elif a[2]=='Golden Globe':
            net_sc=a[1]*coeff
        elif a[2]=='BAFTA Film Award':
            net_sc=a[1]*coeff
        else:
            net_sc=a[1]
        awards_score1.append((int(a[0]),a[1],a[2],net_sc))
    add=0
    count=0
    year=[0]
    aw_score=[]
    for b in list(sorted(awards_score1)):
        add=add+b[3]
        #print(mov_avg,i[1])
        year1=year[-1]
        if(b[0]>year1):
            aw_score.append((b[0],add))
            year1=b[0]
            year.append(year1)
        else:
            del aw_score[-1]
            aw_score.append((b[0],add))
            year1=b[0]
            year.append(year1)
    aw_score=sorted(aw_score)
    #for mo,count in zip(mo_score,range(len(mo_score))):
    #    if int(mo[0])>=my_year:
    #        actor_movie_score=(mo_score[count-1][1])
    #        break
    #    else:
    #        continue
    for awa,count in zip((aw_score),range(len(aw_score))):
        if awa[0]>=my_year:
            actor_award_score=(aw_score[count-1][1])
            break
        else:
            actor_award_score=(aw_score[count][1])
            continue
    if sum_1==0 and sum_2==0:
        return 0
    else:
        return actor_award_score
```

```python
def get_publicity_score(name):
    import re
    name1=name
    actor_se1=imdb.search_for_person(name1)
        #print(actor_se)
        #for i in range(len(actor_se1)):
            #if (actor_se1[i]["name"]==name1):
    actor_id1=actor_se1[0]["imdb_id"]
    url1 = "http://www.imdb.com/name/"+actor_id1
    response1 = requests.get(url1)
    if response1.status_code == 200:
        #print(response.content)
        page_data_soup1 = BeautifulSoup(response1.content,'lxml')
        #print(page_data_soup1.prettify())
    publicity=page_data_soup1.find_all('div',{'id' : 'details-publicity-listings'})
    if len(publicity)==0:
        return 0
    else:
        pub_score=[]
        a=publicity[0].get_text()
        b=a.split('\n')
        f=list(filter(lambda a: a != '', b))
        for i in f:
            j=i.split(' ')
            if ('Biography' in j)==True:
                x=re.findall(r'\d+',i)
                if len(x)>0:
                    y=int(x[0])*10
                    pub_score.append(y)
            elif ('Magazine' in j)==True:
                x=re.findall(r'\d+',i)
                if len(x)>0:
                    y=int(x[0])*5
                    pub_score.append(y)
            else:
                x=re.findall(r'\d+',i)
                #print(x)
                if len(x)>0:
                    y=int(x[0])
                    pub_score.append(y)
        return(sum(pub_score))
```
